```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from bs4 import Comment
from numpy.linalg import norm
```

```python
player_info = pd.read_csv('player_data.csv')
```

```python
player_info = player_info.iloc[::-1]
player_info = player_info.reset_index(drop=True)
player_info
```

```python
def is_valid_player():
    ret = None
    flag = True
    c1 = None
    c2 = None
    while(flag):
        p = input("Enter the name of a player you'd like to compare: ")
        p = p.lower()
        c1 = p
        df = player_info.loc[player_info['Player'] == p]
        if df.empty == False:
            print(f'You entered: {p}')
            flag = False
        elif df.empty:
            print('Please check spelling of player and try again')
    flag = True
    while(flag):
        print("What do you like the most about " + str(p).title() + "?")
        cat = input("\n\t(A) Scoring\n\t(B) Impact\n\t(C) Both\n")
        cat = cat.upper()
        if cat == "A" or cat == "B" or cat == "C":
            flag = False
            if cat == "A":
                c2 = "Scoring"
            elif cat == "B":
                c2 = "Impact"
            elif cat == "C":
                c2 = "Both"
        else:
            print("\nEnter either A, B, or C")
    ret = (c1, c2)
    return ret
```

```python
def compare_players(p1, p2):
    p1 = p1.drop(columns=['Player', 'PG %', 'SG %', 'SF %', 'PF %', 'C %'])
    p1 = p1.reset_index(drop=True)
    #print(p1)
    p2 = p2.drop(columns=['Player', 'PG %', 'SG %', 'SF %', 'PF %', 'C %'])
    p2 = p2.reset_index(drop=True)
    #print(p2)
    
    p1 = np.squeeze(np.asarray(p1))
    
    p2 = np.squeeze(np.asarray(p2))
    
    res = np.dot(p1, p2)/(norm(p1)*norm(p2))
    
    return res
```

```python
def main():
    lst = []
    for i in range(2):
        player, cat = is_valid_player()
        temp = player_info.loc[player_info['Player'] == player]
        #if cat == "Scoring":
        #temp = temp[]
        lst.append(temp)
    return compare_players(lst[0], lst[1])
```

```python
main()
```

```python
df = player_info

names = df['Player']

df = df.drop(columns=['Player', 'PG %', 'SG %', 'SF %', 'PF %', 'C %'])
df = df.reset_index(drop=True)
length = len(names)
zeros = np.zeros((length, length))
sim_matrix = pd.DataFrame(zeros, columns=names)
for i, r1 in df.iterrows():
    #print("i: " + str(i))
    for j, r2 in df.iterrows():
        if i != j:
            
            sim_matrix.iat[i, j] = np.dot(r1, r2)/(norm(r1)*norm(r2))
        else:
            sim_matrix.iat[i, j] = 1
sim_matrix.index = names
sim_matrix
```
